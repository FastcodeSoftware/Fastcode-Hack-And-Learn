```python
# Import necessary libraries
from fastai.vision.all import *
```

## Load dataset

```python
# Create DataLoaders using filenames and a labeling function, resizing each image to 224x224
dls = ImageDataLoaders.from_folder(
    '/content/dataset',
    train='train',
    valid='valid',
    item_tfms=Resize(224),
    bs=2,  # small enough to fit into the valid set
    fail_fast=False
)

dls.show_batch()
```

## Train

```python
# Create a vision learner with ResNet34 and accuracy metric
learn = vision_learner(dls, resnet34, metrics=accuracy)

# Fine-tune the model for 5 epochs (1 frozen + 4 unfrozen)
learn.fine_tune(5)

# Show predictions vs actuals from validation set
learn.show_results()
```
